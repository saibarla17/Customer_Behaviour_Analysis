```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\saiba\Downloads\customer_shopping_behavior.csv")
```

```python
df.head(5)
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
lables=['young_adult','adult','Middle_aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels=lables)
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly':30,
    'Quarterly' : 90,
    'Bi-Weekly' : 15,
    'Annually' : 365,
    'Every 3 Months': 90
}	
    
df['purchases_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
